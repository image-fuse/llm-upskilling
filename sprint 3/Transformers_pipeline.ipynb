# Documentation:


#### 1. Sentiment Analysis:
**Definition**: Sentiment analysis is the process of determining the emotional tone or attitude expressed in a piece of text.<br>
**Application**: It is widely used for analyzing customer reviews, social media sentiments, and feedback.

#### 2. Named Entity Recognition (NER):
D**efinition**: Named Entity Recognition involves identifying and classifying entities (e.g., names of people, places, organizations) in text.<br>
**Application**: Useful in information extraction, language understanding, and content categorization.

#### 3. Question Answering:
**Definition**: Question Answering is the task of automatically providing relevant answers to user queries based on a given context.<br>
**Application**: Commonly used in chatbots, search engines, and virtual assistants.

#### 4. Summarization:
**Definition**: Summarization is the process of generating a concise and coherent summary of a longer text document.<br>
**Application**: Helps in quickly understanding the key points of lengthy articles, documents, or news.

#### 5. Text Generation:
**Definition**: Text generation involves creating new pieces of text, often based on a given prompt or context.<br>
**Application**: Used in creative writing, content creation, and generating human-like responses in chatbots.

#### 6. Zero Shot Classification:
**Definition**: Zero Shot Classification is a type of text classification where a model is trained to classify input text into multiple classes, even if it has never seen examples of those classes during training.
**Application**: Useful when dealing with a large number of classes or constantly evolving categories.

# Pipelines

### Sentiment analysis

```python
from transformers import pipeline

classifier = pipeline("sentiment-analysis")
classifier(
    [
        "I've been waiting for a HuggingFace course my whole life.",
        "I hate this so much!",
    ]
)
```

### Zero-shot classificiation

```python
from transformers import pipeline

classifier = pipeline("zero-shot-classification")
classifier(
    "This is a course about the Transformers library",
    candidate_labels=["education", "politics", "business"],
)
```

### Text Generation

```python
from transformers import pipeline

generator = pipeline("text-generation")
generator("In this course, we will teach you how to")
```

### Named Entity Recognition

```python
from transformers import pipeline

ner = pipeline("ner", grouped_entities=True)
ner("My name is Sylvain and I work at Hugging Face in Brooklyn.")
```

### Question Answering

```python
from transformers import pipeline

question_answerer = pipeline("question-answering")
question_answerer(
    question="Where do I work?",
    context="My name is Sylvain and I work at Hugging Face in Brooklyn",
)
```

### Summarization
